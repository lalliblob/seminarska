```python
import pandas as pd
```

```python
df = pd.read_csv('radio_student_articles.csv') 

df.head()
```

```python
df.info()

df.isnull().sum()

df.describe()
```

```python
avtorice = df['avtor'].value_counts()
print(avtorice)
import matplotlib.pyplot as plt
avtorice.plot(kind='bar', title='število člankov na avtorico')
plt.show() 

# filtrirano število avtorjev z več članki
min_člankov = 10  
filtrirane_avtorice = avtorice[avtorice >= min_člankov]

filtrirane_avtorice.plot(kind='bar', title='število člankov na avtorico (najpogostejše avtorice)')

plt.figure(figsize=(12, 6)) 
filtrirane_avtorice.plot(kind='bar', alpha=0.7, color='skyblue') 

plt.title('število člankov na avtorico (najpogostejše avtorice)')
plt.xlabel('avtorice')
plt.ylabel('število člankov')

plt.xticks(rotation=45)  
plt.tight_layout()  

plt.show()
```

```python
tip_oddaje_seštevek = df['tip_oddaje'].value_counts()
print(tip_oddaje_seštevek)

tip_oddaje_seštevek.plot(kind='bar', title='pojavnost oddaj redakcije za kulturo')

import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))  
tip_oddaje_seštevek.plot(kind='bar', alpha=0.8, color='lightblue')  

# Add title and labels
plt.title('pojavnost Oddaj Redakcije za Kulturo', fontsize=16)
plt.xlabel('tip Oddaje', fontsize=12)
plt.ylabel('število pojavitev', fontsize=12)

plt.xticks(rotation=45, ha='right', fontsize=10)  
plt.tight_layout() 

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
import pandas as pd

df = pd.read_csv('radio_student_articles.csv')  

zdruzeni_podatki = df.groupby(['avtor', 'tip_oddaje']).size().unstack().fillna(0)
print(zdruzeni_podatki)

import matplotlib.pyplot as plt

najpogostejših_n_avtorjev = 8

# Iterate over each type of show (column) to create a plot
for tip_oddaje in zdruzeni_podatki.columns:
    
    najpogostejši_avtorji = zdruzeni_podatki[tip_oddaje].sort_values(ascending=False).head(najpogostejših_n_avtorjev)
   
    plt.figure(figsize=(12, 6))
    najpogostejši_avtorji.plot(kind='bar', alpha=0.8, color='lightblue')

    plt.title(f'število člankov na avtorja za tip oddaje: {tip_oddaje} (Najpogostejši {najpogostejših_n_avtorjev} Avtorji)', fontsize=16)
    plt.xlabel('avtor', fontsize=12)
    plt.ylabel('število člankov', fontsize=12)
    plt.xticks(rotation=45, ha='right')  
    plt.tight_layout()  

    plt.show()
```

```python
import pandas as pd
df = pd.read_csv('radio_student_articles.csv')  
df['datum'] =pd.to_datetime(df['datum'], format='%d. %m. %Y – %H.%M', errors='coerce')

from datetime import datetime, timedelta

zadnjih_trideset_dni = datetime.now() - timedelta(days=30)

df_zadnjih_trideset_dni = df[df['datum'] >= zadnjih_trideset_dni]

print("oblika df_zadnjih_trideset_dni:", df_zadnjih_trideset_dni.shape)
print(df_zadnjih_trideset_dni.head())

pojavnost_tipov_oddaj = df_zadnjih_trideset_dni['tip_oddaje'].value_counts()
print(pojavnost_tipov_oddaj)

import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
pojavnost_tipov_oddaj.plot(kind='bar', color='lightgreen', alpha=0.8)

plt.title('Pojavnost Tipov Oddaj v Zadnjih 30 Dneh', fontsize=16)
plt.xlabel('Tip Oddaje', fontsize=12)
plt.ylabel('Število Pojavov', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 

plt.show()
```
